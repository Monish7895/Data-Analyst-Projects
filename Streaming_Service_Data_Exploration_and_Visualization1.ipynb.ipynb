**Handling the null value in director column**

```python
df["director"].fillna("director not specified",inplace=True)
```

**Handling the null value in cast column**

```python
df["cast"].fillna("cast not specified",inplace=True)
```

**Handling the null value in country column**

```python
df["country"].fillna("unknown country",inplace=True)
```

```python
sns.histplot(data=df,x="release_year",bins=15)
plt.xlabel('release_year')
plt.ylabel('frequency')
plt.title('Distribution of title by year')

plt.show()
```

```python
type_count=df.groupby("type").nunique()["title"].sort_values(ascending=False)
type_count
```

```python
type_count.plot(kind='bar', color=['blue','orange'])
plt.xlabel('Type')
plt.ylabel('count')
plt.title('content Distribution by Type')
plt.xticks(rotation=0)
plt.show()
```

```python
unique_title_rating_count=df.groupby('rating')['title'].nunique()
unique_title_rating_count
```

```python
unique_titles_by_rating_count = df.groupby("rating")["title"].nunique().sort_values(ascending=False)

plt.figure(figsize=(14, 6))
sns.barplot(x=unique_titles_by_rating_count.index, y=unique_titles_by_rating_count.values)
plt.title('Distribution of Titles by Rating')
plt.xlabel('Release Year')
plt.ylabel('Count of Unique Titles')
plt.xticks(rotation=0, ha='right')
plt.show()
```
